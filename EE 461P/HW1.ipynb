# Question 2.a: Maximum likelihood estimate (10 pts)

Suppose a manager at an internet sales company wants to estimate how fast his salesperson is generating successful leads. Instead of recording the time for each lead, the time taken to generate the next 3 leads are recorded, i.e., there is one recording (denoting the elapsed time) for every 3 consecutive leads. For a specific salesperson, the time intervals recorded are {1,3,1.5,4,2,7,1.2,2,4,3.1} hours. 

A statistician suggests that if these time intervals are assumed to arise by i.i.d. sampling from the following distribution:
$$ p(t) = \frac{1}{C \times \theta^{3}}t^{2}exp^{-\frac{t}{\theta}},$$
(where C is a normalizing constant). Therefore, if $\theta$ can be estimated, then he can provide detailed information
about the lead generation process, including average rates, variances etc.

Find the Maximum Likelihood estimate for $\theta$ based on the recorded observations.


## Answer

# 2(a)

$$ p(x, \theta) = \frac{1}{C \times \theta^{3}}x^{2}exp^{-\frac{x}{\theta}}$$

##### We wish to maximize the probability that  $ p(x, \theta)$ generated the observed data

$$\mathscr{L(\theta)}  = \Pi^{k}_{i=1} p(x_i , \theta)$$

$$ln(\mathscr{L(\theta)}) = \Sigma^{k}_{i=1} ln\left(p(x_i | \theta)\right) = \Sigma^{k}_{i=1}\  ln\left(\frac{1}{C \times \theta^{3}}x^{2}exp^{-\frac{x_i}{\theta}}\right) $$

$$ ln(\mathscr{L(\theta)}) = \Sigma^{k}_{i=1}\ 2ln(x) - 3ln(\theta) - \frac{x_i}{\theta}$$
$$ ln(\mathscr{L(\theta)}) = \Sigma^{k}_{i=1}\left[ 2ln(x_i)\right] -\ \Sigma^{k}_{i=1}\left[ 3ln(\theta) \right]\ -\ \Sigma^{k}_{i=1}\frac{x_i}{\theta}$$

##### The first term will be a constant factor, so we can ignore it for our purposes


$$ ln(\mathscr{L(\theta)}) =  -\ \Sigma^{k}_{i=1}\left[ 3ln(\theta) \right]\ -\ \Sigma^{k}_{i=1}\frac{x_i}{\theta}$$

$$ ln(\mathscr{L(\theta)}) =  -\ k\left[ 3ln(\theta) \right]\ -\ \frac{1}{\theta}\Sigma^{k}_{i=1}x_i$$

##### For our data, k = 10, and $\Sigma^{k}_{i=1}x_i$ = 28.8, Thus:

$$ ln(\mathscr{L(\theta)}) =  -30ln(\theta) \ -\ \frac{28.8}{\theta}$$

##### We wish to maximize this function, so we will take the derrivative with respect to $\theta$


$$ \frac{\partial}{\partial\theta}ln(\mathscr{L(\theta)}) =  \frac{-30ln}{\theta} \ +\ \frac{28.8}{\theta^2} = 0$$


$$\rightarrow \theta \approx0.96$$

## Answer


# 2.b.a)


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#Bivariate Gaussian with sigmaX = 4, sigmaY = 3, and rho = 0
def p1(x,y) :
    Sx = 4
    Sy = 3
    return (1/(2*np.pi*Sx*Sy))*np.exp((-(((x*x)/(Sx*Sx)) + ((y*y)/(Sy*Sy))))/2)

#Bivariate Gaussian with sigmaX = 4, sigmaY = 3, and rho = 0.5
def p2(x,y,Sx=None,Sy=None,rho=None) :
    if Sx==None:
        Sx = 4
    if Sy==None:
        Sy = 3
    if rho==None:
        rho = 0.5
    return (1/(2*np.pi*Sx*Sy*(1-(rho)**2)**(0.5)))*np.exp((-( ((x*x)/(Sx*Sx)) + ((y*y)/(Sy*Sy)) - ((2*rho*x*y)/(Sx*Sy)) ) / (2*(1-(rho)**2)) ))
```

```python
xlist = np.linspace(-10.0, 10.0, 100)
ylist = np.linspace(-10.0, 10.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = p1(X,Y)

plt.figure()
cp = plt.contour(X, Y, Z, 10)
plt.clabel(cp, inline=True, fontsize=10)
plt.title('Sigma_x = 4, Sigma_y = 3, rho = 0')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# 2.b.c)

```python
"""
NOTE TO GRADER:
The rotation to our distribution is happening in this function. 

p1 and p2 return the value of the distribution at a given x,y coordinate. 

When a coordinate pair are passed to p3, the value of the distribution
at (x*cos(theta) - y*sin(theta), x*sin(theta) + y*cos(theta)) is returned.
This amounts to a rotation by [theta].

"""

def p3(x1,y1,theta) :
    Sx = 4
    Sy = 3
    rho = 0.5
    
    #Perform rotation by theta
    x = x1*np.cos(theta) - y1*np.sin(theta)
    y = x1*np.sin(theta) + y1*np.cos(theta)
   
    return (1/(2*np.pi*Sx*Sy*(1-(rho)**2)**(0.5)))*np.exp((-( ((x*x)/(Sx*Sx)) + ((y*y)/(Sy*Sy)) - ((2*rho*x*y)/(Sx*Sy)) ) / (2*(1-(rho)**2)) ))
```

```python
theta = -np.pi/4
```

```python

#Generate a grid of points
xlist = np.linspace(-10.0, 10.0, 100)
ylist = np.linspace(-10.0, 10.0, 100)

X,Y = np.meshgrid(xlist,ylist)
Z3 = p3(X,Y,theta)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z3, rstride=5, cstride=5)
ax.view_init(30, 330)
plt.show()
```

The plot shown above and below is the rotated distribution



```python
xlist = np.linspace(-10.0, 10.0, 100)
ylist = np.linspace(-10.0, 10.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = p3(X,Y,theta)

plt.figure()
cp = plt.contour(X, Y, Z, 10)
plt.clabel(cp, inline=True, fontsize=10)
plt.title('Sigma_x = 4, Sigma_y = 3, rho = 0.5, ROTATED by -45*')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.model_selection import ShuffleSplit, cross_val_score, KFold
from sklearn.linear_model import  Ridge, Lasso

#alphas  = np.logspace(-10, 8, num=100, base=2)
alphas = np.linspace(150,160,100)

#c = 155.287878788
#offset = 0.1
#alphas = np.linspace(c+offset,c-offset,100)

Ridge_scores = pd.DataFrame(columns=['alpha','MSE'])

for alpha in alphas:
    cv = KFold(n_splits=5)
    clf = Ridge(alpha=alpha)
    cv_score = cross_val_score(clf, X_train, y_train, cv=cv,scoring='neg_mean_squared_error').mean()
    #print(0 - cv_score)
    Ridge_scores = Ridge_scores.append(pd.DataFrame(data={'alpha': [alpha], 'MSE': [0 - cv_score]}),ignore_index=True)

```

# Interpretation of this error function...

### i) ...in terms of data dependent noise variance

Perhaps you have lower confidence in some of the data points you collected compared to others. You may choose the $r_n$ value for that data point to be small so that if it contributes a large amount of error, you will lessen it's contribution. This could be to reflect that you don't trust the squared error of this prediction as much, i.e: it has higher variance than other data points.

### ii) ...in terms of replicated data points

If you have repeated data points, it may be reasonable to set $r_n $ = 0 for some of them so that they do not over-contribute to your error. Perhaps your dataset it highly skewed such that 99% of your data is identical and what you really care about is your model's ability to correctly identify the 1% of dissimilar data. This type of error function could be useful in this scenario.
